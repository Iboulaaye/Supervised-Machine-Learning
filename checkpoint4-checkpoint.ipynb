we start by importing the libraries to use numpy, panda, seaborn, skitlearn for the regression.

We draw a graph to see the distribution of rooms in the dataset, we see that most houses have between 3 and 5 rooms with a predominance for 3 rooms.
We use the correlation map to see the dependency between the outflow(price) and the inputs. 

```python
df3=house_data[["price","bedrooms","bathrooms","sqft_living","grade","sqft_above","sqft_basement","sqft_living15","lat"]]
df3.head()
```

#2/ Following the map we see that the price depends mainly on : "bedrooms"(0,31%),"bathrooms"(0,51),"sqft_living"(0,7),"grade"(0,61),"sqft_above"(0,67),"sqft_basement"(0,39),"sqft_living15"(0,59),"lat"(0,31)
In the rest of the processing we will use only these outputs, hence df3 which contains these inputs. 


```python
sns.scatterplot(data=df3, x="sqft_living" , y="price")
plt.xlabel("sqft_living")
plt.ylabel("price")
```

Compared to the other inputs, sqft_living has a greater dependence on our output (price) so we will use it when studying our one input linear and polynomial regression. 

```python
x=df3["sqft_living"].values[:,np.newaxis]
y=df3["price"].values
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.10,random_state=10)
```

# 3/
we divide our dataset into training and test data. So we will have 10% for testing and 90% for training. Note also that x takes the input sqft_living as mentioned before and y which is our output remains the price. 

```python
model=LinearRegression()   #build linear regression model
model.fit(x_train,y_train)  #fitting the training data
predicted=model.predict(x_test) #testing our modelâ€™s performance
```

```python
model.coef_
```

```python
# This line allows us to plot our model according to our input, we see that it does not really meet expectations
plt.scatter(x,y,color="b")
plt.title("Linear Regression")
plt.ylabel("price")
plt.xlabel("sqft_living")
plt.plot(x,model.predict(x),color="k")
plt.show()
```

```python
#5/ Measure of the linear regression
print("MSE", mean_squared_error(y_test,predicted))
print("R squared", metrics.r2_score(y_test,predicted))
```

```python
#6/ Multilinear regression
x1=df3[["sqft_living","sqft_above","sqft_basement","sqft_living15","bedrooms","bathrooms","lat"]].values
y1=df3["price"].values
x1_train,x1_test,y1_train,y1_test=train_test_split(x1,y1,test_size=0.10,random_state=10)
```

```python
model=LinearRegression()   #build linear regression model
model.fit(x1_train,y1_train)  #fitting the training data
predicted=model.predict(x1_test) #testing our modelâ€™s performance
```

```python
print("MSE", mean_squared_error(y1_test,predicted))
print("R squared", metrics.r2_score(y1_test,predicted))
```

Linear regression: squared error=0.50
Multilinear regression: squared error=0.59
From these results we see that with the same data for testing and training the multilinear has decreased the loss function which improve the R squared
so the model is more appropriate than the linear regression

```python
#7/ Polynomial regression
x=df3["sqft_living"].values[:,np.newaxis]
y=df3["price"].values

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.10, random_state=10)  #splitting data
lg=LinearRegression()
poly=PolynomialFeatures(degree=2)

x_train_fit = poly.fit_transform(x_train) #transforming our input data
lg.fit(x_train_fit, y_train)
x_test_ = poly.fit_transform(x_test)
predicted = lg.predict(x_test_)
```

```python
plt.scatter(x, y, color="r")
plt.title("Linear regression")
plt.ylabel("sqft_living")
plt.xlabel("price")
plt.plot(x, lg.predict(poly.fit_transform(x)), color="k") 
```

```python
print("MSE: ", metrics.mean_squared_error(y_test, predicted))
print("R squared: ", metrics.r2_score(y_test,predicted))
```

```python
#Multi polynomiale regression 
x=df3[["sqft_living","sqft_above","sqft_basement","sqft_living15","bedrooms","bathrooms","lat"]].values
y=df3["price"].values

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.10, random_state=10)  #splitting data
lg=LinearRegression()
poly=PolynomialFeatures(degree=2)

x_train_fit = poly.fit_transform(x_train) #transforming our input data
lg.fit(x_train_fit, y_train)
x_test_ = poly.fit_transform(x_test)
predicted = lg.predict(x_test_)
```

```python
print("MSE: ", metrics.mean_squared_error(y_test, predicted))
print("R squared: ", metrics.r2_score(y_test,predicted))
```

Linear regression: squared error = 0.50
Multilinear regression: squared error = 0.59
polynomiale regression : squared error = 0.57
Multi polynomiale regression : squared error = 0.67
When we compare the squared errors we see that the one that is closer to 1 is the multi polynomial model so the results to be predicted will be closer to the true expected results.
